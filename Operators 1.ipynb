```python
#1) Arithmatic Operators

#+ : Addition
#- : subtraction
#* : Multiplication
#/ : Division
#** : Exponential
#% : Modulus
#// : Floor Division

#1) Addition:-

x = 10
y = 20
z = x+y
print(z)

x = 10
y = x + 20+20
z = x+y
print(z)

x = 10
y = x + 20+20
z = (x+y) + (x+y)
print(z)

#2) Subtraction:-

a = 20
b = 30
c = b-a
print(c)

a = 20
b = a
c = b-a+10
print(c)

#3) Multiplication :

l = 10
m = 8
n = l*m
print(n)

l = 10
m = l
n = l*m +10
print(n)

l = 10
m = l
n = l+m*10
print(n)

#/ : Division

p = 15
q = 5
r = p/q
print(r)

# Convert float into integer
p = 15
q = 5
r = int(p/q)
print(r)

p = 15
q = p
r = p/q*10+5
print(r)

p = 15
q = p
r = int(p/q+10*5-10)
print(r)

#** : Exponential

x = 2
y = 8
z = x**y
print(z)

x = 2+2
z = x**2
print(z)

#% : Modulus

no1 = 24
no2 = 5
no3 = no1 % no2
print(no3)


m = 44
n = 6
r = m % n
print(r)

no1 = 24
no2 = 5
no3 = no1 % no2 +5 *no2
print(no3)

#// : Floor Division

n1 = 10
n2 = 2
n3 = n1//n2
print(n3)
print(type(n3))

n1 = 10
n2 = 3
n3 = n1//n2
print(n3)
print(type(n3))

#Assigment Operators:--
# = :-- Equal to
# += :-- Addition Equal to
# -= :-- Subtraction Equal to
# *= :-- Multiplication Equal to
# /= :-- Division Equal to
# **= :-- Exponential Equal to
# %= :-- Modulus Equal to
# //= :-- Floor Division Equal to



# += :-- Addition equal to :---

a = 10
c += a #(c = c+a)
print(c)

a = 10
c += a #(c = c+a)
print(c)

a = 10
a += a+a*a
print(a)

# -= :-- Subtraction Equal to

a = 10
c -= a #(c = c-a)
print(c)

a = 10
a -= a-a*a
print(a)

# *= :-- Multiplication Equal to

nos5 = 5
nos4 = 10
nos5 *= nos4
print(nos5)


# /= :-- Division Equal to


nos5 = 5
nos4 = 10
nos5 /= nos4
print(nos5)

#**= :-- Exponential Equal to

nos5 = 2
nos4 = 5
nos5 **= nos4
print(nos5)


# %= :-- Modulus Equal to

nos1 = 3
nos2 = 5
nos2 %= nos1
print(nos2)

# //= :-- Floor Division Equal to

nos1 = 3
nos2 = 5
nos2 //= nos1
print(nos2)

```
